```python
import os
import pandas as pd
from pathlib import Path
```

```python
PROJECT_ROOT = Path(os.getenv("PROJECT_ROOT"))
```

```python
df = pd.read_csv(PROJECT_ROOT / "data/intermediate/data_all.csv")
df
```

```python
def min_max_scale(column):
    min_val = column.min()
    max_val = column.max()
    return (column - min_val) / (max_val - min_val)
```

```python
numeric_columns = ['temperature', 'rainfall', 'humidity', 'population_density', 'population', 'vehicles_count', 'roads_length']

for column in numeric_columns:
    df[column] = min_max_scale(df[column])

df
```

```python
df.to_csv(PROJECT_ROOT / "data/intermediate/data_all_norm.csv", index=False)
```
